#Data Types and Structure Questions

1. What are data structures, and why are they important?
   - Data structures are ways of organizing and storing data so it can be used efficiently.
They are important because they affect how fast and easily you can store, search, update, and manage data in a program.

2. Difference between mutable and immutable data types (with examples)
   - Mutable: Can be changed after creation
Examples: list, dictionary, set
 immutable: Cannot be changed after creation
  Examples: tuple, string, int

3. Main differences between lists and tuples
   - Lists are mutable, tuples are immutable
   Lists use [], tuples use ()
   Tuples are generally faster and safer for fixed data

4. How dictionaries store data
   - Dictionaries store data as key–value pairs Keys are unique and mapped to values using hashing.

5. Why use a set instead of a list?
   - Sets automatically remove duplicates,
Faster membership testing (in),
Useful when order doesn’t matter

6. What is a string, and how is it different from a list?
   - A string is a sequence of characters and is immutable.
A list is a sequence of items and is mutable.

7. How do tuples ensure data integrity?
   - Because tuples cannot be changed, data inside them stays safe from accidental modification.

8. What is a hash table, and how does it relate to dictionaries?
   - A hash table is a data structure that stores data using hashes.
Python dictionaries are implemented using hash tables, allowing fast lookups.

9. Can lists contain different data types?
   - Yes

10. Why are strings immutable in Python?
    - Immutability makes strings:
More memory-efficient,
Safer to use,
Faster to process (they can be reused internally)

11. Advantages of dictionaries over lists
    - Faster lookups by key,
More readable data,
Ideal for structured data (e.g., records)

12. How do sets handle duplicate values?
    - Sets automatically remove duplicates:
  nums = {1, 2, 2, 3}
  Result: {1, 2, 3}

13. When is a tuple preferable over a list?
    - When data should not change (e.g., coordinates),
When using data as dictionary keys,
When performance matters

14. How does the in keyword differ for lists and dictionaries?
    - List: checks values
Dictionary: checks keys

15. Can you modify elements of a tuple?
    - No. Tuples are immutable, so their elements cannot be changed after creation.

16. What is a nested dictionary (use case)?
    - A dictionary inside another dictionary

17. Time complexity of dictionary access
    - Average case: O(1) (constant time)
This makes dictionaries very fast.

18. When are lists preferred over dictionaries?
    - When order matters,
When working with simple collections,
When data is accessed by position (index)

19. Why are dictionaries considered unordered?
    - Traditionally, dictionaries did not store items in order.
Modern Python preserves insertion order, but retrieval is still key-based, not position-based.

20. Difference between a list and a dictionary in data retrieval
    - List: retrieved by index (list[0])
Dictionary: retrieved by key (dict["key"])
